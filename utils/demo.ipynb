```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
x = np.linspace(0.2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 4, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
yy[:, 0]
```

```python
fig1, f1_axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True)
x = np.arange(0, 10, 0.005)
y = np.exp(-x/2.) * np.sin(2*np.pi*x)
f1_axes[0, 0].plot(x, y)
fig2 = plt.figure(constrained_layout=True)
spec2 = gridspec.GridSpec(nrows=2, ncols=2, figure=fig2)
f2_ax1 = fig2.add_subplot(spec2[0, 0])
f2_ax1.plot([0, 1, 2])
f2_ax2 = fig2.add_subplot(spec2[0, 1])
f2_ax2.bar([0, 1, 2], [2, 3, 4])
f2_ax3 = fig2.add_subplot(spec2[1, 0])
f2_ax4 = fig2.add_subplot(spec2[1, 1])
```

```python
fig3 = plt.figure(figsize=[18, 10], constrained_layout=True)
gs = fig3.add_gridspec(4, 8)
f3_ax1 = fig3.add_subplot(gs[:, :4])
f3_ax1.plot([1, 2, 3])
f3_ax1.legend('o')
fig3.add_subplot(gs[:2, 4:6])
fig3.add_subplot(gs[:2, 6:8])
fig3.add_subplot(gs[2:, 4:6])
fig3.add_subplot(gs[2:, 6:8])
```
